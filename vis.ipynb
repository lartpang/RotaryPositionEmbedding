```python
import torch
from RoPE import RoPE1D
import matplotlib.pyplot as plt
```

```python
rope_1d = RoPE1D(dim=40).cuda()
image = torch.ones(1, 100, 40, dtype=torch.float32, device='cuda')
image = rope_1d(image)
q = image
k = image

weight = (q @ k.transpose(-2, -1)) / q.shape[-1]  # b,hwl,hws
weight = weight[0].cpu()

plt.imshow(weight)
plt.show()
```

```python
hwl, hws = weight.shape
mid_sim = weight[hwl // 2, :]
plt.plot(mid_sim)
plt.show()
```
